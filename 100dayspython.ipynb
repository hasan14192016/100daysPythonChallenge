### Day 20: Expense Tracker Project

```python
expenses = []

def add_expenses():
  category = input("Enter the expense category (e.g. Food, Rent, Entertainment): ")
  amount = float(input("Enter the amount: $"))
  expenses.append((category,amount))

def view_expenses():
  if not(expenses):
    print("No expenses avaialable...")
    return

  total = 0
  category_totals = {}

  for category,amount in expenses:
    total += amount
    category_totals[category] = category_totals.get(category,0)+ amount
  print("\nTotal Expenses: $",total)
  print("Breakdown by category:")
  for category,amount in category_totals.items():
    print(f'{category}: ${amount:.2f}.')

def expense_tracker():
  while True:
    print("\n1.Add Expense \n2.View Expense \n3.Exit")
    choice = input("choose an option: ")

    if choice == '1':
      add_expenses()
    elif choice == '2':
      view_expenses()
    elif choice == '3':
      break
    else:
      print("Bad choice input!!!!")


expense_tracker()
```

### Day 21: Unit Converter Project

```python
def cel_to_feh(x):
  return (x*9/5) + 32

def feh_to_cel(x):
  return (x-32) * 5 / 9

def kilo_to_meter(x):
  return x *1000

def meter_to_kilo(x):
  return x/1000

value = float(input("Enter the amount: "))
unitfrom = input("Enter the unit you want to transform from (K: Kilometer, M: Meter, F: Feherinheit, C: Cellilose) : ").lower()

if unitfrom == 'k':
  print(kilo_to_meter(value))
elif unitfrom == 'm':
  print(meter_to_kilo(value))
elif unitfrom == 'f':
  print(feh_to_cel(value))
elif unitfrom == 'c':
  print(cel_to_feh(value))
else:
  print("Invalid unit!!!!")
```

### Day 30: Palindrome Checker

```python
def is_palindrome(text):
  cleaned = ""
  for char in text:
    if char.isalnum():
      cleaned += char.lower()

  reversed_text = cleaned[::-1]
  return text.lower() == reversed_text

text = input("Statement: ")
if is_palindrome(text):
  print("Palindrome")
else:
  print("Non-palindrome")
```

### Day 42: Class Methods, Static Methods, & Instance Methods

```python
######Class Methods
class Person:
  total_persons = 0
  def __init__(self,name,age):
    self.name = name
    self.age = age
    Person.total_persons += 1

  def greet(self):
    print(f"Hello {self.name}.")

  @classmethod
  def get_total_people(P):
    return P.total_persons

Hasan = Person("Hasan",45)
Hasan.greet()
Badr = Person("Badr",34)
Badr.greet()
print(f"Total People = {Person.get_total_people()}")
```

```python
######static Methods
class Person:
  total_persons = 0
  def __init__(self,name,age):
    self.name = name
    self.age = age
    Person.total_persons += 1

  def greet(self):
    print(f"Hello {self.name}.")

  @classmethod
  def get_total_people(P):
    return P.total_persons

  @staticmethod
  def is_even(age):
    return age % 2 == 0

Hasan = Person("Hasan",45)
Hasan.greet()
Badr = Person("Badr",34)
Badr.greet()
print(f"Total People = {Person.get_total_people()}")
print(Person.is_even(9))
print(Person.is_even(16))
```

```python
class BankAccount:
  active_accounts = 0
  def __init__(self,holder,balance,number):
    self.account_holder = holder
    self.balance = balance
    self.account_number = number
    BankAccount.active_accounts += 1

  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance -= amount
      print("Money has been withdrawn")
    else:
      print("No sufficient money")

  @classmethod
  def get_active_acoounts_number(cls):
    return cls.active_accounts

  @staticmethod
  def is_valid_account_number(acc_number):
    return len(acc_number) == 10

  def deposit(self,amount):
    self.balance += amount
    print(f"{amount} riyals has been added to your balance.")

a = BankAccount("Hasan",1000,"12345566")
b = BankAccount("Badr",100,"1234556")
a.deposit(1000)
b.withdraw(1000)
a.withdraw(500)
print(BankAccount.get_active_acoounts_number())
print(BankAccount.is_valid_account_number("123654789"))
print(BankAccount.is_valid_account_number("1023654789"))

```

```python
def is_leap_year(year):
  if year %4 == 0:
    if year % 100 == 0:
      if year %400 == 0:
        return True
      else:
        return False
    else:
      return True
  else:
    return False

year_to_check = int(input("Year:\n"))
if is_leap_year(year_to_check):
  print(f"{year_to_check} is a leap year")
else:
  print(f"{year_to_check} is not a leap year")
```

### Day 44: Prime Numbers Project

```python
import math
def is_prime(n):
  if n<1:
    return False
  else:
    for i in range(2,int(math.sqrt(n))+1):
      if n%i == 0:
        return False
    return True

number = int(input("Number:\n"))
if is_prime(number):
  print(f"{number} is prime")
else:
  print(f"{number} is not prime")
```

### Day 52: Christmas Secret Santa

```python
import random

def assign_secret_santa(participants):
  random.shuffle(participants)
  print("Inside secret santa\n",participants)
  santa_pairs = {}
  for i in range(len(participants)):
    santa_pairs[participants[i]] = participants[(i+1) % len(participants)]

  return santa_pairs

participants = []
print("Enter the names of participants. write 'Done' when you finish...")

while True:
  name = input("Name: ")
  if name.lower() == 'done':
    break

  participants.append(name)

if len(participants)<2:
  print("Not enough participants. Please Add least 2 paticipants")
else:
  santa_pairs = assign_secret_santa(participants)

  print("\n---Secret Santa Pairs---")
  for santa,recipient in santa_pairs.items():
    print(f"{santa} will give a gift to {recipient}")

  with open("secret_santa_pairs.txt",'w') as file:
    file.write("---Secret Santa Pairs---\n")
    for santa,recipient in santa_pairs.items():
      file.write(f"{santa} will give a gift to {recipient}\n")

  print("\nPairing saved to secret_santa_pairs.txt")
```

### Day 54: Snow Animations

```python
import pygame
import random

pygame.init()

WIDTH,HEIGHT = 800,600
screen = pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption("Snow Animation")

class snowflake:
  def __init__(self):
    self.x = random.randint(0,WIDTH)
    self.y = random.randint(-HEIGHT,0)
    self.size = random.randint(2,5)
    self.speed = random.uniform(1,3)

  def fall(self):
    self.y += self.speed
    if self.y > HEIGHT:
      self.y = random.randint(-50,-10)
      self.x = random.randint(0,WIDTH)
      self.size = random.randint(2,5)
      self.speed = random.uniform(1,3)

  def draw(self,screen):
    pygame.draw.circle(screen,(255,255,255),(self.x,self.y),self.size)

num_snowflakes = 200
snowflakes = [snowflake() for _ in range(num_snowflakes)]

running = True

clock = pygame.time.Clock()

while running:
  screen.fill((30,30,60))

  for event in pygame.event.get():
    if event.type == pygame.QUIT:
      running = False

  for snowflake in snowflakes:
    snowflake.fall()
    snowflake.draw()

  pygame.display.flip()
  clock.tick(30)

pygame.quit()
```

```python
def generate_strong_password(length):
  if length<4:
    return "password length must be at least 4"
  lower = random.choice(string.ascii_lowercase)
  upper = random.choice(string.ascii_uppercase)
  digit = random.choice(string.digits)
  special_character = random.choice(string.punctuation)
  remaining_length = length - 4
  remaining_characters = ''.join(random.choice(string.ascii_letters+string.digits+string.punctuation) for i in range(remaining_length))
  password = lower + upper + digit +special_character + remaining_characters
  password = ''.join(random.sample(password,len(password))) #shuffle the characters

  return password

password_length = int(input("Enter the length of the password: "))
print(f"THe produced password is {generate_strong_password(password_length):>20}")

```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Tic-Tac_Toc")

current_player = 'X'
buttons =[[None,None,None],[None,None,None],[None,None,None]]
game_over = False

def highlight_winner(b1,b2,b3):
  b1.config(bg='lightgreen')
  b2.config(bg='lightgreen')
  b3.config(bg='lightgreen')

def check_winner():
  global game_over

  for row in range(3):
    if buttons[row][0]["text"] ==  buttons[row][1]["text"] == buttons[row][2]["text"] != "":
      highlight_winner(buttons[row][0],buttons[row][1],buttons[row][2])
      messagebox.showinfo("Game Over",f"Player {buttons[row][0]['text']} wins!")
      game_over = True
      return

  for col in range(3):
    if buttons[0][col]["text"] ==  buttons[1][col]["text"] == buttons[2][col]["text"] != "":
      highlight_winner(buttons[0][col],buttons[1][col],buttons[2][col])
      messagebox.showinfo("Game Over",f"Player {buttons[0][col]['text']} wins!")
      game_over = True
      return

  if buttons[0][0]['text'] == buttons[1][1]["text"] == buttons[2][2]["text"] != "":
      highlight_winner(buttons[0][0],buttons[1][1],buttons[2][2])
      messagebox.showinfo("Game Over",f"Player {buttons[0][0]['text']} wins!")
      game_over = True
      return

  if buttons[0][2]['text'] == buttons[1][1]["text"] == buttons[2][0]["text"] != "":
    highlight_winner(buttons[0][2],buttons[1][1],buttons[2][0])
    messagebox.showinfo("Game Over",f"Player {buttons[0][2]['text']} wins!")
    game_over = True
    return

  if all(buttons[row][col]["text"]!="" for row in range(3) for col in range(3) ) and not game_over:
    messagebox.showinfo("Game Over","It's a draw!!!")
    game_over = True
    return

def button_clicks(row,col):
  global current_player
  if buttons[row][col]["text"] =="" and not game_over:
    buttons[row][col]['text'] = current_player
    buttons[row][col].config(fg="black" if current_player=='X' else "blue")

    check_winner()

    current_player = "O" if current_player == 'X' else 'X'

def reset_game():
  global current_player,game_over
  current_player = "X"
  game_over = False
  for row in range(3):
    for col in range(3):
      buttons[row][col]["text"] = ""
      buttons[row][col].config(bg="SystemButtonFace")

for row in range(3):
  for col in range(3):
    buttons[row][col] = tk.Button(root,text = "",width=10,height= 3,font=('Helvetica',20),command=lambda r=row,c=col: button_clicks(r,c))
    buttons[row][col].grid(row=row,column = col)

reset_button = tk.Button(root,text = "Reset",font=("Helvetica",15),command = reset_game)
reset_button.grid(row=3,column = 0,columnspan=3)


root.mainloop()
```

### Day 61 : Fireworks Animation

```python
import pygame
import random
import math
```

```python
pygame.init()

width,height = 800,600
screen = pygame.display.set_mode((width,height))
pygame.display.set_caption("New Year's Firework")

clock = pygame.time.Clock()

class Particle:
  def __init__(self,x,y,color):
    self.x = x
    self.y = y
    self.color = color
    self.radius = random.randint(3,6)
    self.angle = random.uniform(0,2* math.pi)
    self.speed = random.uniform(2,8)
    self.gravity = 0.2

  def move(self):
    self.x += math.cos(self.angle) * self.speed
    self.y += math.sin(self.angle) * self.speed
    self.speed -= 0.1
    self.y += self.gravity
    self.radius -= 0.1
    self.radius = max(0,self.radius)

  def draw(self,surface):
    if self.radius > 0:
      pygame.draw.circle(surface,self.color,(int(self.x),int(self.y)),int(self.radius))

class Fireworks:
  def __init__(self,x,y):
    self.x = x
    self.y = y
    self.color =random.choice([(255,0,0),(0,255,0),(0,0,255),(255,255,0),(255,0,255),(0,255,255)])
    self.particles = [Particle(self.x,self.y,self.color) for _ in range(50)]
    self.exploded = False

  def explode (self):
    for particle in self.particles:
      particle.move()
    self.particles = [p for p in self.particles if p.radius>0]
    if not self.particles:
      self.exploded = True

  def draw(self,surface):
    for particle in self.particles:
      particle.draw(surface)

#Main game loop
def main():
  fireworks = []
  running = True
  while running:
    screen.fill((0,0,0))

    for event in pygame.event.get():
      if event.type == pygame.QUIT:
        running = False

    if random.randint(1,20) == 1:
      fireworks.append(Fireworks(random.randint(100,width-100),random.randint(100,height//2)))

    for firework in fireworks:
      firework.explode()
      firework.draw(screen)

    fireworks = [fw for fw in fireworks if not fw.exploded]

    pygame.display.flip()

    clock.tick(60)

  pygame.quit()

if __name__ == "__main__":
  main()
```

### Day 66 : Emirp Numbers

```python
def is_prime(num):
  if num <= 1:
    return False
  for i in range(2,int(num ** 0.5) + 1):
    if num % i == 0:
      return False
  return True

def reverse_number(num):
  return int(str(num)[::-1])

def is_emirp(num):
  if not is_prime(num):
    return False
  reversed_num = reverse_number(num)
  if num != reversed_num and is_prime(reversed_num):
    return True
  return False

number = int(input("Number to check: "))
if is_emirp(number):
  print(f"{number} is an emirp>>>")
else:
  print(f"{number} is not an emirp....")

```

### Day 79: ATM Machine Project

```python
user_account ={
  "pin":"1234",
  "balance": 1000
}

def display_menu():
  print("\nWelcome to the virtual ATM.")
  print("1. Check balance.")
  print("2. Deposit money.")
  print("3. Withdraw money.")
  print("4. Exit")

def check_balance():
  print(f"\nYour balance: ${user_account['balance']}")

def deposit():
  try:
    amount = float(input("\nENter the money amount you want to deposti: "))
    if amount>0:
      user_account["balance"] += amount
      print(f"\n${amount} has benn added to your account.")
    else:
      print("Invalid input.Please enter a positive amount.")
  except ValueError:
    print("Invalid input.please enter a valid number")

def withdraw():
  try:
    amount = float(input("\nENter the money amount you want to withdraw: "))
    if amount>0:
      if amount<= user_account["balance"]:
        user_account["balance"] -= amount
        print(f"\n${amount} has benn taken from your account.\nYour new balance is ${user_account['balance']}")
      else:
        print("Invalid input. your balance is not sufficient.")
    else:
      print("Invalid input.Please enter a positive amount.")

  except ValueError:
    print("Invalid input.please enter a valid number")

def atm_machine():
  print("Welcome to the ATM Machine")
  attempts = 3
  while attempts>0:
    entered_pin = input("Your PIN: ")
    if entered_pin == user_account["pin"]:
      print("Login successfully....")

      while True:
        display_menu()
        choice = input("Your choice: ")

        if choice == "1":
          check_balance()
        elif choice == "2":
          deposit()
        elif choice == "3":
          withdraw()
        elif choice == "4":
          print("\nTHank you for using our ATM System...")
          break
        else:
          print("INvalid input. Please try again....")
      break
    else:
      attempts -= 1
      print(f"Incorrect PIN. You have {attempts} tries left...")
  if attempts == 0:
    print("\nToo many incorrect attempts.Exiting program....")

if __name__ == "__main__":
  atm_machine()
```

```python
# merge 2 dictionaries and add the values if a key is common between the dictionaries
dic1 = {1:10,2:20}
dic2 = {2:30,3:40}
new_dict = dic1.copy()

for key,value in dic2.items():
  if key in dic1.keys():
    new_dict[key] += value
  else:
    new_dict[key] = value

print(new_dict)
```

```python
nested = [1,[2,3],[4,[5,6]],7]
```

```python
def flatten_list(nested_list):
  flat_list =[]
  for item in nested_list:
    if isinstance(item,list):
      flat_list.extend(flatten_list(item))
    else:
      flat_list.append(item)
  return flat_list

print(flatten_list(nested))
```

```python
def is_anagram(str1,str2):
  return  sorted(str1) == sorted(str2)

print(is_anagram("silent","listen"))
```

```python
def second_largest(lst):
  unique_list = list(set(lst))
  print(sorted(unique_list)[1])

second_largest([1,5,2,7,4,1.5])
```

```python
from collections import Counter

def most_frequent(lst):
  count = Counter(lst)
  return count.most_common(1)[0][0]

print(most_frequent([1,3,2,2,1,3,4,5,3,2,2,1,3,4,1,1,2,2,3,3,2]))
```

```python
import string

def is_pangrams(s):
  return set(string.ascii_lowercase).issubset(set(s.lower()))

print(is_pangrams("The quick brown fox jamps over the lazy dog"))
print(is_pangrams("Hello,world"))
```

```python
def longest_word(sentence):
  words = sentence.split()
  return(max(words,key=len))

print(longest_word("If multiple words have the same length, return the first one"))
```

```python
def list_intersection(lst1,lst2):
  return list(set(lst1) & set(lst2))

print(list_intersection([1,2,3,4],[3,4,5,6]))
```
